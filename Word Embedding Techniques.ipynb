## word Embedding Techniques using Embedding Layer in Keras

```python
import tensorflow as tf
tf.__version__
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
##sentences
sent = ['the weather is beautiful', 
        'the weather is rainy', 
        'I will have a cup of coffee', 
        'the food was great', 
       'I am practicing hard to become a good coder',
       'Data Science is not that hard']
```

```python
#vocabulary size
voc_size= 5000
```

```python
#Onehot representation
onehot_rep= [one_hot(words, voc_size) for words in sent]
onehot_rep
```

Word Embedding representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length= 10
embedded_doc=pad_sequences(onehot_rep, padding='post',maxlen=sent_length)
embedded_doc
```

```python
#feature representation dimension
dim = 10
model=Sequential()
model.add(Embedding(voc_size, dim, input_length=sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
embedded_doc[0]
```

```python
print(model.predict(embedded_doc))
```
